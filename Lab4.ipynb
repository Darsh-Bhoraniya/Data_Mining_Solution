### 1) Write a python program to compute distance between Given two objects represented by the tuples (22, 1, 42, 10) and (20, 0, 36, 8): 
(a) Compute the Euclidean distance between the two objects. <BR>
(b) Compute the Manhattan distance between the two objects. <BR>
(c) Compute the Minkowski distance between the two objects, using p = 3. <BR>
(d) Compute the supremum distance between the two objects.<BR>


```python
import math as m
x = (22, 1, 42, 10)
y = (20, 0, 36, 8)

#Euclidean distance:
sum_euc = 0
for i in range(0, len(x)):
    sum_euc += (x[i] - y[i])**2
print(f"Euclidean distance : {m.sqrt(sum_euc)}")

#Manhatten distance:
sum_man = 0
for i in range(0, len(x)):
    sum_man += abs(x[i] - y[i])
print(f"Manhatten distance : {sum_man}")

#Minkowski distance:
sum_min = 0
p = 3
for i in range(0, len(x)):
    sum_min += (abs(x[i] - y[i]))**p
print(f"Minkowski distance : {sum_min**(1/p)}")

#Supremum distance:
sup_max = 0
for i in range(len(x)):
    sup_max = max(sup_max, abs(x[i] - y[i]))
print(f"Superenum distance : {sup_max}")
```

### 3) Write programs to perform the following tasks of preprocessing.

Equal Width Binning <BR>
Equal Frequency/Depth Binning

```python
# Equal width binning

import pandas as pd
import numpy as np

# Sample data
data = [5, 10, 11, 13, 15, 35, 50, 55, 72, 92, 204, 215]

# Convert data to a pandas DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Define the number of bins
num_bins = 3

# Calculate the bin edges
bin_edges = np.linspace(df['Value'].min(), df['Value'].max(), num_bins + 1)

# Perform Equal Width Binning
df['Equal_Width_Bin'] = pd.cut(df['Value'], bins=bin_edges, labels=False, include_lowest=True)

print("Equal Width Binning:\n", df)
```

### 4)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
# min-max normalization on the 'Age' column.
# Min-max normalization is one of the most common ways to normalize data.
# For every feature, the minimum value of that feature gets transformed into a 0, 
# the maximum value gets transformed into a 1, and every other value gets transformed into a decimal between 0 and 1.
df5 = df4
new_max = 10
new_min = 1
df5['New Age'] = ((df4['Age'] - df4['Age'].min()) / (df4['Age'].max() - df4['Age'].min())) * (new_max - new_min) * new_min
df5
```

```python
# Get A Minimum Age
min = df4[df4['Age']>=1]['Age'].min()
min
```

```python
# Get A Maximum Age
max = df4['Age'].max()
max
```

```python
# Create the new column new age using decimal scaling of age attribute
df6 = df4
df6['New Age'] = df4['Age']/10**len(str(int(df4['Age'].max())))
df6
```

```python
# Z-score normalization of all the ages into the column Age-Zcore
df6['Age-Zscore'] = (df6['Age'] - df6['Age'].mean()) / df6['Age'].std()
df6
```

```python
# Print the co-relation
df6[['Age', 'New Age', 'Age-Zscore']].corr()
```
